```
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```
# input features
x = torch.tensor([[73, 67, 43],
                  [91, 88, 64],
                  [87, 134, 58],
                  [102, 43, 37],
                  [69, 96, 70]],
                  dtype=torch.float)

print(x)
```

```
# target values
y = torch.tensor([[56, 70],
                  [81, 101],
                  [119, 133],
                  [22, 37],
                  [103, 119]],
                 dtype=torch.float)

print(y)
```

```
# weights
w = torch.randn(3, 2, requires_grad=True)
print(w)
```

```
# bias
b = torch.randn(1, 2, requires_grad=True)
print(b)
```

```
# learning rate
alpha = 0.00001
```

```
# training for 100 iterations

loss_values = []

for i in range(100):
  z = torch.mm(x, w) + b
  e = (z-y)**2
  l = e.mean() # mse loss
  loss_values += [l.item()]
  l.backward() # gradient calculation
  with torch.no_grad(): # for updating the leaf variables
    w -= alpha*w.grad
    b -= alpha*b.grad
    w.grad.zero_()
    b.grad.zero_()

plt.plot(np.array(loss_values), 'r')
```

```
preds = torch.mm(x, w) + b
mse = ((preds-y)**2).mean() # final mse
print(mse)
```
